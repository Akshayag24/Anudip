```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load the dataset
df = pd.read_csv("lung_cancer_dataset_sample.csv")
df.head()   # display the first 5 rows for preview
```

```python
df.info()   ##Shows column names, data types, and non-null values
```

```python
df.isnull().sum()   ## checks for missing data in each column
```

```python
df.describe()
```

```python
## Data Visualization
```

```python
## countplot- gender vs diagnosis
sns.countplot(data=df, x='Gender', hue='Lung_Cancer_Diagnosis', palette='Set2')
plt.title("Gender vs Lung Cancer Diagnosis")
plt.xlabel("Gender")
plt.ylabel("Patient Count")
plt.legend(title="Diagnosis")
plt.show()
```

```python
## Histogram- Age distribution
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='purple')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
## heatmap- Cough vs diagnosis
cough_heat = pd.crosstab(df['Cough'], df['Lung_Cancer_Diagnosis'])
sns.heatmap(cough_heat, annot=True, cmap='Blues', fmt='d')
plt.title("Cough vs Lung Cancer Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Cough (Yes/No)")
plt.show()
```

```python
## grouped Bar plot- weightloss vs diagnosis
weight_group = df.groupby(['Weight_Loss', 'Lung_Cancer_Diagnosis']).size().reset_index(name='Count')
sns.barplot(data=weight_group, x='Weight_Loss', y='Count', hue='Lung_Cancer_Diagnosis', palette='coolwarm')
plt.title("Weight Loss vs Diagnosis")
plt.xlabel("Weight Loss (Yes/No)")
plt.ylabel("Number of Patients")
plt.legend(title="Diagnosis")
plt.show()
```

```python
- **Cough and weight loss** are more common in diagnosed patients.
- ** smoking ** appears to be a strong risk factor.
- Most patients are above ** 50 years ** of age.
- ** Gender ** distribution shows both males and females are affected, with possible variation.
- ** visual analytics ** can help in symptom-based risk identification.
```
